### Violin Plot ###

```python
import pandas
import plotly.express as px
```

```python
import numpy as np
d1 = np.random.uniform (0,1,5000)
d2 = np.random.normal (0,1,5000)
d31 = np.random.normal (0,5,500)
d32= np.random.normal (20,5,1000)
d3=np.hstack([d31,d32])
#d3 = np.ones (100)
d4= np.random.geometric(p=0.5, size=100)
d5 = np.random.exponential (size=100)
```

```python
fig1=px.violin (d1, box=True, points='all')
fig1.show()
fig1=px.violin (d2, box=True, points='all')
fig1.show()
fig1=px.violin(d3, box=True, points='all')
fig1.show()
fig1=px.violin(d4, box=True, points='all')
fig1.show()
fig1=px.violin(d5, box=True, points='all')
fig1.show()
```
