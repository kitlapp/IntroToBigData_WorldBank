# Matrix Decomposition #
## non negative matrix decomposition

Matrix **A**, of dimensions **m x n** can broken into two submatrices:
- **W**  of dimensions **m x k** 
- **H**  of dimensions **k x n**
- k<<m,n known as **number of latent factores**




https://www.geeksforgeeks.org/machine-learning/non-negative-matrix-factorization/

Thus matrix a can be ``broken'' into W and H, which when multiplied they can be approximate the original matrix.
### A (m×n) ≈ W (m×k) x H (k×n)
### to find W and H you try to minimize the function

$$min (\sum_{i,j,k}(A(i,j) - W(i,k)xH(k,j)))$$

### Application of matrix decomposition ###

Assume:
-   m- users
-   n- products 


| Customer | Product A | Product B | Product C | Product D |
| -------- | --------- | --------- | --------- | --------- |
| C1       | 3         | 0         | 1         | 0         |
| C2       | 5         | 1         | 0         | 0         |
| C3       | 0         | 2         | 4         | 3         |
| C4       | 0         | 1         | 3         | 4         |


```python
import numpy as np
from sklearn.decomposition import NMF
```

```python
V = np.array([[3, 	0, 	1, 	0],[5, 	1, 	0, 	0],[0, 	2, 	4, 	3],[0, 	1, 	3, 	4]])
print (V.shape)
V
```

```python

model = NMF(n_components=2, init='random', random_state=0)
W = model.fit_transform(V)
H = model.components_
print ('error=',model.reconstruction_err_)
```

```python
print ('customer–latent feature matrix W')
print (W)
print ('\n')
print ('latent feature-product matrix  H')
print (H)
```
